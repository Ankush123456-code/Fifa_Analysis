### heighest Pay salary player

```python
playe_salary=df[['short_name',"wage_eur"]]
```

```python
playe_salary.head()
```

```python
playe_salary=playe_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
playe_salary.head()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(playe_salary["short_name"])[0:5],list(playe_salary['wage_eur'])[0:5],color=['Black',"r",'g',"pink",'orange'])
```
